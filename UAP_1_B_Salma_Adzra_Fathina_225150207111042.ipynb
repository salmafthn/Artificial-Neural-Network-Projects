## Implementasi Backpropagation

```python
import numpy as np
import time


# Referensi Implementasi
class BPNN:
    def __init__(self, layer_conf, learn_rate=0.1, max_epoch=10000, max_error=0.1, print_per_epoch=100):
        self.layer_conf = layer_conf
        self.learn_rate = learn_rate
        self.max_epoch = max_epoch
        self.max_error = max_error
        self.print_per_epoch = print_per_epoch
        self.weights = None

    @staticmethod
    def _sig(x):
        return 1 / (1 + np.exp(-x))

    @staticmethod
    def _sigd(x):
        s = BPNN._sig(x)
        return s * (1 - s)

    def fit(self, X, target):
        start_time = time.time()
        np.random.seed(1)

        nin = [np.empty(i) for i in self.layer_conf]
        n = [np.empty(j + 1) if i < len(self.layer_conf) - 1 else np.empty(j)
             for i, j in enumerate(self.layer_conf)]
        self.weights = [np.random.rand(self.layer_conf[i] + 1, self.layer_conf[i + 1])
                for i in range(len(self.layer_conf) - 1)]
        dw = [np.empty((self.layer_conf[i] + 1, self.layer_conf[i + 1])) for i in range(len(self.layer_conf) - 1)]
        d = [np.empty(s) for s in self.layer_conf[1:]]
        din = [np.empty(s) for s in self.layer_conf[1:-1]]

        epoch = 0
        mse = 1

        for i in range(0, len(n) - 1):
            n[i][-1] = 1

        while (self.max_epoch == -1 or epoch < self.max_epoch) and mse > self.max_error:
            epoch += 1
            mse = 0

            for r in range(len(X)):
                n[0][:-1] = X[r]

                for L in range(1, len(self.layer_conf)): # Forward Pass
                    nin[L] = np.dot(n[L - 1], self.weights[L - 1])
                    n[L][:len(nin[L])] = self._sig(nin[L])

                e = target[r] - n[-1] # MSE
                mse += sum(e ** 2)

                d[-1] = e * self._sigd(nin[-1]) # Backprop
                dw[-1] = self.learn_rate * d[-1] * n[-2].reshape((-1, 1))

                for L in range(len(self.layer_conf) - 1, 1, -1):
                    din[L - 2] = np.dot(d[L - 1], np.transpose(self.weights[L - 1][:-1]))
                    d[L - 2] = din[L - 2] * np.array(self._sigd(nin[L - 1]))
                    dw[L - 2] = (self.learn_rate * d[L - 2]) * n[L - 2].reshape((-1, 1))

            for L in range(len(self.weights)): # Update Bobot GD
                    self.weights[L] += dw[L]

            mse /= len(X)

            if self.print_per_epoch > -1 and epoch % self.print_per_epoch == 0:
                print(f'Epoch {epoch}, MSE: {mse}')

        execution = time.time() - start_time
        print("Execution Time: {:.2f} seconds".format(execution))
        return self

    def predict(self, X):
        n = [np.empty(len(i)) for i in self.weights]
        nin = [np.empty(len(i[0])) for i in self.weights]
        predict = []
        n.append(np.empty(len(self.weights[-1][0])))

        for x in X:
            n[0][:-1] = x
            for L in range(0, len(self.weights)):
                nin[L] = np.dot(n[L], self.weights[L])
                n[L + 1][:len(nin[L])] = self._sig(nin[L])
            predict.append(n[-1].copy())

        return predict
```

```python
class BPNN_SGD(BPNN):
    def fit(self, X, target):
        start_time = time.time()
        np.random.seed(1)

        nin = [np.empty(i) for i in self.layer_conf]
        n = [np.empty(j + 1) if i < len(self.layer_conf) - 1 else np.empty(j)
             for i, j in enumerate(self.layer_conf)]
        self.weights = [np.random.rand(self.layer_conf[i] + 1, self.layer_conf[i + 1])
                for i in range(len(self.layer_conf) - 1)]
        dw = [np.empty((self.layer_conf[i] + 1, self.layer_conf[i + 1])) for i in range(len(self.layer_conf) - 1)]
        d = [np.empty(s) for s in self.layer_conf[1:]]
        din = [np.empty(s) for s in self.layer_conf[1:-1]]

        epoch = 0
        mse = 1

        for i in range(0, len(n) - 1):
            n[i][-1] = 1

        while (self.max_epoch == -1 or epoch < self.max_epoch) and mse > self.max_error:
            epoch += 1
            mse = 0

            for r in range(len(X)):
                n[0][:-1] = X[r]

                for L in range(1, len(self.layer_conf)):  # Forward Pass
                    nin[L] = np.dot(n[L - 1], self.weights[L - 1])
                    n[L][:len(nin[L])] = self._sig(nin[L])

                e = target[r] - n[-1]  # MSE
                mse += sum(e ** 2)

                d[-1] = e * self._sigd(nin[-1])  # Backprop
                dw[-1] = self.learn_rate * d[-1] * n[-2].reshape((-1, 1))

                for L in range(len(self.layer_conf) - 1, 1, -1):
                    din[L - 2] = np.dot(d[L - 1], np.transpose(self.weights[L - 1][:-1]))
                    d[L - 2] = din[L - 2] * np.array(self._sigd(nin[L - 1]))
                    dw[L - 2] = (self.learn_rate * d[L - 2]) * n[L - 2].reshape((-1, 1))

                for L in range(len(self.weights)):  # Update Bobot SGD
                    self.weights[L] += dw[L]

            mse /= len(X)

            if self.print_per_epoch > -1 and epoch % self.print_per_epoch == 0:
                print(f'Epoch {epoch}, MSE: {mse}')

        execution = time.time() - start_time
        print("Execution Time: {:.2f} seconds".format(execution))
        return self
```

```python
from sklearn.metrics import f1_score, confusion_matrix

gd = BPNN(layer_conf=[13, 5, 1], learn_rate=0.1, max_epoch=1000, max_error=0.01, print_per_epoch=100)
gd.fit(X_train, y_train)
y_pred_gd = np.round(gd.predict(X_test)).astype(int)
```

```python
sgd = BPNN_SGD(layer_conf=[13, 5, 1], learn_rate=0.1, max_epoch=1000, max_error=0.01, print_per_epoch=100)
sgd.fit(X_train, y_train)
y_pred_sgd = np.round(sgd.predict(X_test)).astype(int)
```

```python
# Gunakan data yang telah dipersiapkan sebelumnya untuk melatih model Backpropagation dengan SGD dan lakukan evaluasi terhadap akurasi
# Ujikan kedua model GD dan SGD untuk kemudian dibandingkan
# Jika dirasa perlu, model dapat dilanjutkan dievaluasi menggunakan F1-Score dan Confusion Matrix (Opsional)

from sklearn.metrics import f1_score, confusion_matrix, accuracy_score

accuracy_gd = accuracy_score(y_test, y_pred_gd)
f1_gd = f1_score(y_test, y_pred_gd)
cm_gd = confusion_matrix(y_test, y_pred_gd)
print(f"Akurasi GD: {accuracy_gd:.2f}")
print(f"F1-Score GD: {f1_gd:.2f}")
print("Confusion Matrix GD:\n", cm_gd)

accuracy_sgd = accuracy_score(y_test, y_pred_sgd)
f1_sgd = f1_score(y_test, y_pred_sgd)
cm_sgd = confusion_matrix(y_test, y_pred_sgd)
print(f"Akurasi SGD: {accuracy_sgd:.2f}")
print(f"F1-Score SGD: {f1_sgd:.2f}")
print("Confusion Matrix SGD:\n", cm_sgd)
```
